### Percentiles: 
The percentiles to include in the output. All should fall between 0 and 1. </br>

The default is [.25, .5, .75], which returns the 25th, 50th, and 75th percentiles.

eg.
          count   mean         std    min    25%    50%    75%    max
GOOG      2.0    160.0   56.568542  120.0  140.0  160.0  180.0  200.0


max-min = 80

80*25%= 20

0%   : min + 0%(max-min)  = 120 : min

25%  : min + 25%(max-min) = 140 

50%  : min + 50%(max-min) = 160

75%  : min + 75%(max-min) = 180

100% : min + 100%(max-min)= 200 : max

### Merging data frames

```python
# Creating 3 data frames

df1 = pd.DataFrame(    {'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3'],
                        'C': ['C0', 'C1', 'C2', 'C3'],
                        'D': ['D0', 'D1', 'D2', 'D3']},
                        index=[0, 1, 2, 3])

df2 = pd.DataFrame(    {'A': ['A4', 'A5', 'A6', 'A7'],
                        'B': ['B4', 'B5', 'B6', 'B7'],
                        'C': ['C4', 'C5', 'C6', 'C7'],
                        'D': ['D4', 'D5', 'D6', 'D7']},
                         index=[4, 5, 6, 7])

df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                        'B': ['B8', 'B9', 'B10', 'B11'],
                        'C': ['C8', 'C9', 'C10', 'C11'],
                        'D': ['D8', 'D9', 'D10', 'D11']},
                        index=[8,9,10,11])

print(df1)
print('-'*68, sep='')
print(df2)
print('-'*68, sep='')
print(df3)
```

#### Concatenation with axis=0

```python
df_cat1 = pd.concat([df1,df2,df3], axis=0)
print("\nAfter concatenation along row\n",'-'*30, sep='')
print(df_cat1)
```

#### Concatenation with axis=1

```python
df_cat2 = pd.concat([df1,df2,df3], axis=1)
print("\nAfter concatenation along column\n",'-'*60, sep='')
print(df_cat2)
```

#### fillna()

```python
df_cat2.fillna(value=0)
print("\nAfter filling missing values with zero\n",'-'*60, sep='')
print(df_cat2.fillna(value=0))
```

### Merging by single common key

```python
left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
   
right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']})

print("\nThe DataFrame 'left'\n",'-'*30, sep='')
print(left)

print("\nThe DataFrame 'right'\n",'-'*30, sep='')
print(right)
```

```python
merge1= pd.merge(left,right, on='key')
print("\nAfter simple merging with 'inner' method\n",'-'*50, sep='')
print(merge1)
```

### Merging by 2 common keys

```python
# create DF

left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                        'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3']})
    
right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                               'key2': ['K0', 'K0', 'K0', 'K0'],
                                  'C': ['C0', 'C1', 'C2', 'C3'],
                                  'D': ['D0', 'D1', 'D2', 'D3']})

print("\nThe DataFrame 'left'\n",'-'*30, sep='')
print(left)

print("\nThe DataFrame 'right'\n",'-'*30, sep='')
print(right)
```

```python
pd.merge(left, right, on=['key1', 'key2'])

# Include, if left.(key1, key2) == right.(key1, key2) else exclude 
```

#### Left Merge

```python
print("\nThe DataFrame 'left'\n",'-'*30, sep='')
print(left)

print("\nThe DataFrame 'right'\n",'-'*30, sep='')
print(right)

pd.merge(left, right, how='left',on=['key1', 'key2'])
```

#### Right Merge

```python
pd.merge(left, right, how='right',on=['key1', 'key2'])
```

### Joins

```python
#join operators
import pandas as pd
left = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']},
                      index=['K0', 'K1', 'K2']) 

right = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                    'D': ['D0', 'D2', 'D3']},
                      index=['K0', 'K2', 'K3'])

print("\nThe DataFrame 'left'\n",'-'*30, sep='')
print(left)

print("\nThe DataFrame 'right'\n",'-'*30, sep='')
print(right)
```

```python
left.join(right)
```

```python
right.join(left)
```

```python
print("\nThe DataFrame 'left'\n",'-'*30, sep='')
print(left)

print("\nThe DataFrame 'right'\n",'-'*30, sep='')
print(right)

left.join(right, how='outer')
```

#### Apply user defined function

```python
# Define a function
def testfunc(x):
    if (x> 5):
        return (10*x)
    else:
        return (x/10)
```

```python
df['UserFuncApplied'] = df['col1'].apply(testfunc)
df
```

#### Apply internal function

```python
df['InternalFuncApplied']= df['col3'].apply(len)
df
```

#### Apply lamda function

```python
# a = lambda x: np.sqrt(x)

import numpy as np
df['LamdaFuncApplied'] = df['UserFuncApplied'].apply(lambda x: np.sqrt(x))
df
```

## Date & Time

### Create a Range of Dates

```python
import pandas as pd
print(pd.date_range('1-1-2011', periods=5)) #DDMMYY
```

```python
start = pd.datetime(2011, 1, 1) # YY , MM, DD
end = pd.datetime(2011, 1, 5)
print(pd.date_range(start, end))
```

### Change the Date Frequency

```python
print(pd.date_range('1/1/2011', periods=5,freq='M'))
```

### bdate_range
bdate_range() stands for business date ranges. Unlike date_range(), it excludes Saturday and Sunday.

```python
print(pd.bdate_range('1/1/2011', periods=7))
```

### Time
By passing a string literal, we can create a timedelta object.

```python
print(pd.Timedelta('2 Days 2 h 15 min 30 sec'))
```

### Operations on Time & Date
You can operate on Series/ DataFrames and construct timedelta64[ns] Series through subtraction operations on datetime64[ns] Series, or Timestamps.

Let us now create a DataFrame with Timedelta and datetime objects and perform some arithmetic operations on it 

```python
s = pd.Series(pd.date_range('2012-1-1', periods=3, freq='D'))
print(s)
td = pd.Series([ pd.Timedelta(hours=i) for i in range(3) ])
print(td)

df = pd.DataFrame(dict(A = s, B = td))
df
```

### Add dates

```python
s = pd.Series(pd.date_range('2012-1-1', periods=3, freq='D'))
td = pd.Series([ pd.Timedelta(days=i) for i in range(3) ])
# print(s)
print(td)
df = pd.DataFrame(dict(A = s, B = td))
print(df)
df['C']=df['A']+df['B']
print(df.describe())
df
```

### read new csv

```python
url = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/US_Crime_Rates/US_Crime_Rates_1960_2014.csv"
crime = pd.read_csv(url)
crime.head()
```

### Convert the type of the column Year to datetime

```python
print(crime.Year)
crime.Year = pd.to_datetime(crime.Year, format='%Y')
crime.info()
crime.head()
```

### Set the Year column as the index of the dataframe

```python

crime = crime.set_index('Year', drop = True)
crime.head()
```

### Group the year by decades and sum the values

```python
# Uses resample to sum each decade
crimes = crime.resample('10AS').sum()
print(crimes.iloc[:5,:5])
```

### Uses resample to get the max value only for the "Population" column

```python

population = crime['Population'].resample('10AS').max()
print(population.head())

# s = [1,2,3,4,5]
# print(s.max())

# Updating the "Population" column
crimes['Population'] = population
crime.head()
```

#### Motivation
- Useful for All - AI/ML Engineer, Data Scientist, Data Analyst, Business Analyst
  
- Our brains are **wired to process visual information** much faster than numbers or text.

- We can **instantly recognize patterns, trends, shapes, and outliers** in images or charts.

- Large tables of numbers take **more time and mental effort** to interpret.



#### Introduction
- **Data visualization** turns raw numbers into pictures.

- It helps **detect patterns, trends, and outliers**.

- It guides **feature selection and model building** in machine learning.

- And most importantly, it **communicates insights clearly** to managers and stakeholders.

- There are so many library options for Data Visualizations - Matplotlib, Seaborn, Pandas

  
| Feature                   | Pandas    | Matplotlib            | Seaborn                 |
| ------------------------- | --------- | --------------------- | ----------------------- |
| Ease of use               | High      | Medium                | High                    |
| Customization             | Low       | Very High             | Medium                  |
| Aesthetics                | Basic     | Customizable          | Beautiful default       |
| Works well with DataFrame | Yes       | No (needs conversion) | Yes                     |

- If you want to plot a column from a DataFrame, you usually need to extract it as a Series or a list/NumPy array first.

#### Data Visualization in Pandas
- Pandas has a built-in .plot() method for Series and DataFrames.

- Under the hood, Pandas uses Matplotlib as its backend, but you don’t need to import or call it manually.

- Best for quick exploratory data analysis (EDA).

### Synthetic Dataset: Customer Shopping Behavior for Online Store

#### Features

- **CustomerID** → Unique identifier  
- **Age** → Integer (18–65)  
- **Gender** → Categorical (Male, Female)  
- **Annual_Income (k)** → Numeric (100–1200)  
- **Spending_Score** → Numeric (1–100, like Mall dataset)  
- **Membership** → Categorical (Silver, Gold, Platinum)  
- **Purchase_Frequency** → Numeric (1–20 per month)  
- **Preferred_Category** → Categorical (Electronics, Clothing, Grocery, Sports, Beauty)  
- **Churn** → Binary target (0 = Active, 1 = Left store → good for classification tasks)  


````python
# import required modules and packages
import pandas as pd
import numpy as np
````

````python
# For reproducibility
np.random.seed(42)

# Number of samples
n = 500
````

````python
# Generate synthetic data
data = {
    "CustomerID": range(1, n+1),
    "Age": np.random.randint(18, 66, n),
    "Gender": np.random.choice(["Male", "Female"], n),
    "Annual_Income_k": np.random.randint(100, 1200, n),
    "Spending_Score": np.random.randint(1, 101, n),
    "Membership": np.random.choice(["Silver", "Gold", "Platinum"], n, p=[0.5, 0.3, 0.2]),
    "Purchase_Frequency": np.random.randint(1, 21, n),
    "Preferred_Category": np.random.choice(
        ["Electronics", "Clothing", "Grocery", "Sports", "Beauty"], n
    ),
    "Churn": np.random.choice([0, 1], n, p=[0.8, 0.2])  # 20% churn rate
}
````

````python
df = pd.DataFrame(data)
df.head()
````

````python
df.info()
````

In the context of **data visualization and modeling**:

- **X variables** are called **independent variables**, **predictors**, or **features**.  
- **Y variable** is called the **dependent variable**, **target**, or **response**.


````python
# important method
df['Churn'].value_counts()
````

### 2. Bar Plot

**Bar Plot** - Gender distribution, Membership counts, Preferred_Category frequencies.

````python
# Count of Membership types
df["Membership"].value_counts()
````

````python
# Vertical Bar Plot => figsize=(width, height) => width & height are in inches.
df["Membership"].value_counts().plot(kind="bar", title="Membership Distribution", figsize=(5, 3))
````

````python
df["Membership"].value_counts().plot.bar(title="Membership Distribution", figsize=(5, 3))
````

````python
# Vertical Bar Plot
df["Preferred_Category"].value_counts().plot(kind="barh", title="Preferred Category Distribution", figsize=(5, 3))
````

````python
# Grouped Bar Plot
df.groupby(["Membership", "Gender"])["Spending_Score"].mean().unstack()
````

````python
# Grouped Bar Plot
df.groupby(["Membership", "Gender"])["Spending_Score"].mean().unstack().plot(kind="bar", title="Avg Spending Score", figsize=(5, 3))
````

````python
# Stacked Bar Plot
df.groupby(["Membership", "Gender"])["Spending_Score"].mean().unstack().plot(kind="bar", stacked=True, title="Stacked Bar Plot", figsize=(5, 3))
````

### 3. Histogram Plot

A histogram is used to show the distribution of numeric data by splitting it into bins.
```python
df["column"].plot(kind="hist")
# or
df["column"].plot.hist()
```

````python
# Basic Histogram - Shows distribution of a single numeric column.
df["Age"].plot(kind="hist", bins=10, title="Age Distribution", rwidth=0.8, figsize=(5, 3))
# bins=10 -> number of intervals
# rwidth=0.8 -> gap between bars
````

````python
# Histogram with Density Curve - Use density=True to show probabilities instead of raw counts.
df["Annual_Income_k"].plot(kind="hist", bins=15, density=True, alpha=0.6, title="Annual Income Distribution", figsize=(5, 3))
````

````python
# Histogram by Category (Group-wise) => We can overlay histograms for subgroups.
df[df["Gender"]=="Male"]["Spending_Score"].plot(kind="hist", bins=15, alpha=0.5, label="Male")
df[df["Gender"]=="Female"]["Spending_Score"].plot(kind="hist", bins=15, alpha=0.5, label="Female")

# %matplotlib inline
import matplotlib.pyplot as plt

plt.legend()
plt.title("Spending Score by Gender")
plt.show()
````

### 4. Box Plot

````python
# Basic Boxplot (Single Column) - Shows the spread of a single numeric column.
df["Age"].plot.box( title="Age Distribution", figsize=(5, 3))
````

````python
# Multiple Boxplots (Multiple Numeric Columns) - Compare distributions of several numeric features side by side.
df[["Age", "Spending_Score"]].plot.box( title="Multiple Features", figsize=(5,3))
````

````python
# Boxplot Grouped by a Category => Show how distributions differ across categories.
df.plot.box(column="Spending_Score", by="Gender", figsize=(6,3))
````

````python
# Horizontal Boxplot
df[["Age", "Spending_Score"]].plot(kind="box", vert=False, title="Horizontal Boxplot", figsize=(5,3))
````

### 5. Kde or Density plots

A density plot is a smoothed version of a histogram that estimates the probability distribution of a numeric variable.
```python
df["col"].plot(kind="density")
# or
df["col"].plot.density()

```

````python
!pip install scipy
````

````python
# Basic Density Plot (Single Column) - Shows distribution of one numeric feature.
df["Age"].plot.density( title="Density Plot: Age", figsize=(10,3))
````

````python
# Multiple Density Plots (Overlayed) - Compare distributions of multiple features.
df[["Age", "Annual_Income_k", "Spending_Score"]].plot.density( title="Density Plot: Multiple Features", figsize=(10,3))
````

````python
# Multiple Density Plots (Overlayed) - Compare distributions of multiple features.
df[["Age", "Annual_Income_k", "Spending_Score"]].plot.density( title="Density Plot: Multiple Features", logx =True, figsize=(10,3))
````

````python
# Density Plot by Category (Overlayed) - Compare distribution across groups (e.g., Male vs Female).
df[df["Gender"]=="Male"]["Spending_Score"].plot(kind="density", label="Male", figsize=(10,3))
df[df["Gender"]=="Female"]["Spending_Score"].plot(kind="density", label="Female", figsize=(10,3))

plt.legend()
plt.title("Spending Score Density by Gender")
plt.show()
````

````python
# log style
df.groupby("Gender")["Spending_Score"].plot.density( legend=True, title="Spending Score by Gender", figsize=(10,3))
````

````python
# Customizations
df[["Age", "Annual_Income_k", "Spending_Score"]].plot.density( title="Density Plot: Multiple Features", logx =True, style="g--",figsize=(10,3))
````

### 6. Area plots

- An area plot is basically a line plot with the area below the line filled in, often used to show trends over time or cumulative proportions.

````python
df["Spending_Score"].head(20).plot(kind="area", title="Area Plot: Spending Score (First 20 Customers)",figsize=(10,3))
````

### 7. Scatter plots

- Scatter plots are among the most important tools in EDA (Exploratory Data Analysis) because \
  they help you visualize relationships between two numeric variables.
- Scatter plot requires numeric columns for the x and y axes. These can be specified by the x and y keywords.

````python
df.plot.scatter(x="Annual_Income_k", y="Spending_Score", title="Income vs Spending Score",figsize=(10,3))
````

````python
# Bigger bubbles = higher spending score.
df.plot.scatter(
    x="Age", y="Annual_Income_k", s=df["Spending_Score"]*3,
    alpha=0.5, title="Income vs Age (Bubble Size = Spending Score)",figsize=(10,4)
)
````

````python
x = np.random.normal(loc=50, scale=15, size=10000)
y = x * 0.8 + np.random.normal(loc=0, scale=10, size=10000)
df_large = pd.DataFrame({"X": x, "Y": y})
df_large.head()
````

````python
df_large.plot.scatter(x="X", y="Y", alpha=0.3, s=10, title="Scatter Plot (10k points)",figsize=(8,4))
````

### 8. Hexagonal bin plots

- A hexbin plot is like a scatter plot, but instead of plotting each point individually, \
  the data space is divided into hexagonal bins and colored by how many points fall into each bin.
- It’s especially useful when you have lots of points (dense scatter), where overlapping makes a scatterplot unreadable.

````python
# Hexbin Plot (better for density)
df_large.plot.hexbin(x="X", y="Y", gridsize=40, cmap="viridis", title="Hexbin Plot",figsize=(8,4))
````

````python
df.plot.hexbin(
    x="Annual_Income_k", y="Spending_Score", 
    gridsize=25,  # number of hexagons across x-axis
    title="Income vs Spending Score (Hexbin)",figsize=(8,4)
)
````

### 9. Pie plots

- If your data includes any NaN, they will be automatically filled with 0. 
- A ValueError will be raised if there are any negative values in your data.
- Pie Chart -> Proportion of Preferred Categories.

````python
df['Gender'].value_counts()
````

````python
df['Gender'].value_counts().plot.pie(
    autopct='%1.1f%%',  # Means a floating-point number with 1 digit before the decimal and 1 digit after the decimal
    title="Customer Gender Distribution",
    figsize=(4,4)
)
````

````python
df['Membership'].value_counts().plot.pie(
    autopct='%1.1f%%', # Means a floating-point number with 1 digit before the decimal and 1 digit after the decimal
    figsize=(4,4),
    title="Customer Membership Distribution"
)
````

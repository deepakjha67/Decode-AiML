#### Motivation
- Useful for All - AI/ML Engineer, Data Scientist, Data Analyst, Business Analyst
  
- Our brains are **wired to process visual information** much faster than numbers or text.

- We can **instantly recognize patterns, trends, shapes, and outliers** in images or charts.

- Large tables of numbers take **more time and mental effort** to interpret.



#### Introduction
- **Data visualization** turns raw numbers into pictures.

- It helps **detect patterns, trends, and outliers**.

- It guides **feature selection and model building** in machine learning.

- And most importantly, it **communicates insights clearly** to managers and stakeholders.

- There are so many library options for Data Visualizations - Matplotlib, Seaborn, Pandas

  
| Feature                   | Pandas    | Matplotlib            | Seaborn                 |
| ------------------------- | --------- | --------------------- | ----------------------- |
| Ease of use               | High      | Medium                | High                    |
| Customization             | Low       | Very High             | Medium                  |
| Aesthetics                | Basic     | Customizable          | Beautiful default       |
| Works well with DataFrame | Yes       | No (needs conversion) | Yes                     |

- If you want to plot a column from a DataFrame, you usually need to extract it as a Series or a list/NumPy array first.

#### Data Visualization in Pandas
- Pandas has a built-in .plot() method for Series and DataFrames.

- Under the hood, Pandas uses Matplotlib as its backend, but you don’t need to import or call it manually.

- Best for quick exploratory data analysis (EDA).

### Synthetic Dataset: Customer Shopping Behavior for Online Store

#### Features

- **CustomerID** → Unique identifier  
- **Age** → Integer (18–65)  
- **Gender** → Categorical (Male, Female)  
- **Annual_Income (k)** → Numeric (100–1200)  
- **Spending_Score** → Numeric (1–100, like Mall dataset)  
- **Membership** → Categorical (Silver, Gold, Platinum)  
- **Purchase_Frequency** → Numeric (1–20 per month)  
- **Preferred_Category** → Categorical (Electronics, Clothing, Grocery, Sports, Beauty)  
- **Churn** → Binary target (0 = Active, 1 = Left store → good for classification tasks)  


````python
# import required modules and packages
import pandas as pd
import numpy as np
````

````python
# For reproducibility
np.random.seed(42)

# Number of samples
n = 500
````

````python
# Generate synthetic data
data = {
    "CustomerID": range(1, n+1),
    "Age": np.random.randint(18, 66, n),
    "Gender": np.random.choice(["Male", "Female"], n),
    "Annual_Income_k": np.random.randint(100, 1200, n),
    "Spending_Score": np.random.randint(1, 101, n),
    "Membership": np.random.choice(["Silver", "Gold", "Platinum"], n, p=[0.5, 0.3, 0.2]),
    "Purchase_Frequency": np.random.randint(1, 21, n),
    "Preferred_Category": np.random.choice(
        ["Electronics", "Clothing", "Grocery", "Sports", "Beauty"], n
    ),
    "Churn": np.random.choice([0, 1], n, p=[0.8, 0.2])  # 20% churn rate
}
````

````python
# convert data to a datframe
df = pd.DataFrame(data)
df.head()
````

````python
# dataframe info
df.info()
````

In the context of **data visualization and modeling**:

- **X variables** are called **independent variables**, **predictors**, or **features**.  
- **Y variable** is called the **dependent variable**, **target**, or **response**.


````python
# Target column value counts
df['Churn'].value_counts()
````

### 4. Box Plot

- A box plot visualizes the five-number summary - `min, Q1, median, Q3, max and outliers`
- Box Plot flags outliers beyond the whiskers (typically 1.5 × IQR).

![Box Plot Example](boxplot.png)


```python
Lower Limit = Q1 - 1.5*IQR
Upper Limit = Q3 + 1.5*IQR
```

````python
data = pd.Series([5,15,16,17,18,19,10,30])
data.plot.box( title="Random Data", figsize=(5,3))
````

### 5. Kde or Density plots

A density plot is a smoothed version of a histogram that estimates the probability distribution of a numeric variable.
```python

# or
df["col"].plot.density()

```

### 8. Hexagonal bin plots

- A hexbin plot is like a scatter plot, but instead of plotting each point individually, \
  the data space is divided into hexagonal bins and colored by how many points fall into each bin.
- It’s especially useful when you have lots of points (dense scatter), where overlapping makes a scatterplot unreadable.

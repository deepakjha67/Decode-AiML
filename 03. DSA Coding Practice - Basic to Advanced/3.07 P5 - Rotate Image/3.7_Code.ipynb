# P5: Rotate Image

```python
"""
Problem : Rotate Image
You are given an n x n 2D matrix representing an image, rotate the image by 90 degrees (clockwise).
"""
```

### Approach 1. Brute-Force

```python
from typing import List
```

```python
"""
To be Implemented - lengthy code
"""
def rotate_bruteforce(matrix):
    n = len(matrix)
    result = [[0]*n for i in range(n)]

    for i in range(n): # row index
        for j in range(n): # column index
            # (i,j) = (j,n-1-i)
            result[j][n-1-i] = matrix[i][j]

    return result

# Time Complexity - O(n^2)
# Space Complexity - O(n^2)
```

```python
rotate_bruteforce([[1,2,3],[4,5,6],[7,8,9]]) # [[7,4,1],[8,5,2],[9,6,3]]
```

### Approach 2. Optimal

```python
"""
To be Implemented
"""
def optimal_rotate(matrix):
    n = len(matrix)

    # transpose the matrix
    for i in range(n): # row index
        for j in range(i+1,n): # column index
            matrix[i][j], matrix[j][i] = matrix[j][i], matrix[i][j]

    # reverse each rows
    for row in matrix:
        row.reverse()

    return matrix

# Time Complexity - O(n^2)
# Space Complexity - O(1)
```

```python
optimal_rotate([[1,2,3],[4,5,6],[7,8,9]]) # [[7,4,1],[8,5,2],[9,6,3]]
```
